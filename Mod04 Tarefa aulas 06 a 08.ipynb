````python
#1) Seu código aqui 
url_sp = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'
r = requests.get(url_sp)
data_sp = r.json()

# ao que vejo, a diferença considerável é que este possui em sua url o estado a ser consultado, de modo que, suponho, podemos definir uma função para
# interagir com os dados desta maneira:
````

````python
def consulta_tesouro(uf='SP', tipo_interessado='Estado'):
    r = requests.get(f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}')
    data = r.json()
    df = pd.DataFrame(data['items'])
    df_filtrado = df[df['tipo_interessado'] == tipo_interessado]
    return df_filtrado
````

````python
df_mg = consulta_tesouro(uf='MG',tipo_interessado='Estado')
print(df_mg)
````
